## Supervised learning: predicting an output variable from high-dimensional observations

### Nearest neighbor and the curse of dimensionality

```python
import numpy as np
from sklearn import datasets

iris = datasets.load_iris()
iris_X = iris.data
iris_y = iris.target

np.unique(iris_y)
```

```python
# Split iris data in train and test data
# A random permutation, to split the data randomly

np.random.seed(0)
indices = np.random.permutation(len(iris_X))

iris_X_train = iris_X[indices[:-10]]
iris_y_train = iris_y[indices[:-10]]
iris_X_test  = iris_X[indices[-10:]]
iris_y_test  = iris_y[indices[-10:]]

# Create and fit a nearest-neighbor classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(iris_X_train, iris_y_train) 
```

```python
knn.predict(iris_X_test)
```

```python
iris_y_test
```

```python
from sklearn.metrics import r2_score
r2_score(knn.predict(iris_X_test), iris_y_test)  # R2スコア
```

```python
figsize(10, 7)

from matplotlib.colors import ListedColormap
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

iris_X = iris.data[:, :2]  # iris_X to two-dem dataset

knn= KNeighborsClassifier()
knn.fit(iris_X, iris_y)

h = .02
x0_min, x0_max = iris_X[:, 0].min() - 1, iris_X[:, 0].max() + 1
x1_min, x1_max = iris_X[:, 1].min() - 1, iris_X[:, 1].max() + 1
x0, x1 = np.meshgrid(np.arange(x0_min, x0_max, h), np.arange(x1_min, x1_max, h))
Z = knn.predict(np.c_[x0.ravel(), x1.ravel()])  # ravelで1次元に変換し，c_で両者を連結したデータに対し予測

Z = Z.reshape(x0.shape)  # x0と同じshapeになるよう，1次元の結果をreshape
pcolormesh(x0, x1, Z, cmap=cmap_light)

scatter(iris_X[:, 0], iris_X[:, 1], c=iris_y, cmap=cmap_bold)
title("3-Class classification")

xlim(xmin=x0_min, xmax=x0_max)
ylim(ymin=x1_min, ymax=x1_max)
```

### Linear model: from regression to sparsity

#### Linear regression

```python
# 糖尿病データセット
diabetes = datasets.load_diabetes()
diabetes_X_train = diabetes.data[:-20]
diabetes_X_test  = diabetes.data[-20:]
diabetes_y_train = diabetes.target[:-20]
diabetes_y_test  = diabetes.target[-20:]
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(diabetes_X_train, diabetes_y_train)
```

```python
# coefficients
regr.coef_
```

```python
# MSE
np.mean((regr.predict(diabetes_X_test) - diabetes_y_test) ** 2)
```

```python
regr.score(diabetes_X_test, diabetes_y_test) 
```

#### Shrinkage

```python
# 次元の呪い
# 各次元毎にデータ数が少ないと，ノイズの影響が大きくなる

figsize(7, 5)

X = np.c_[ .5, 1].T
y = [.5, 1]

test = np.c_[ 0, 2].T
regr = linear_model.LinearRegression()

np.random.seed(0)
for _ in range(6): 
    this_X = .1*np.random.normal(size=(2, 1)) + X
    regr.fit(this_X, y)
    plot(test, regr.predict(test)) 
    plot(this_X, y, 'o')
```

```python
# 学習時の次元数が多い場合，回帰係数をゼロ近くまで縮めるという方法が有効に効く
# そのような手法の一つにリッジ回帰がある(→ Tikhonov正則化)

regr = linear_model.Ridge(alpha=.1)

np.random.seed(0)
for _ in range(6): 
    this_X = .1*np.random.normal(size=(2, 1)) + X
    regr.fit(this_X, y)
    plot(test, regr.predict(test)) 
    plot(this_X, y, 'o')
```

```python
# 誤差が最小になるように alpha を最小化する
alphas = np.logspace(-4, -1, 6)
print(alphas)
print([regr.set_params(alpha=alpha).fit(diabetes_X_train, diabetes_y_train,).score(diabetes_X_test, diabetes_y_test) for alpha in alphas]) 
```

#### Sparsity

```python
# 次元の呪い問題を解決するために，情報量の大きな素性のみを選択することは有効である．
# リッジ回帰では，それぞれの寄与は小さくなるが，ゼロとはしない．

#  もうひとつのアプローチとして，Lasso (least absolute shrinkage and selection operator) がある
# これはいくつかの係数に確実にゼロを当てはめる手法で，パラメータの数が膨大かつそのほとんどが 0 である場合
# 有効に機能することが多い
# これは，スパース法と呼ばれるアプローチである

regr = linear_model.Lasso()

scores = [regr.set_params(alpha=alpha).fit(diabetes_X_train, diabetes_y_train).score(diabetes_X_test, diabetes_y_test) for alpha in alphas]
best_alpha = alphas[scores.index(max(scores))]
regr.alpha = best_alpha

regr.fit(diabetes_X_train, diabetes_y_train)
```

#### Using kernels

```python
figsize(7, 5)

# dataset
X = np.c_[(.4, -.7), (-1.5, -1), (-1.4, -.9), (-1.3, -1.2), (-1.1, -.2), (-1.2, -.4), (-.5, 1.2), (-1.5, 2.1), (1, 1), (1.3, .8),
          (1.2, .5), (.2, -2), (.5, -2.4), (.2, -2.3), (0, -2.7), (1.3, 2.1)].T
Y = [0] * 8 + [1] * 8

# fit the model
for kernel in ('linear', 'poly', 'rbf'):
    clf = svm.SVC(kernel=kernel, gamma=2)
    clf.fit(X, Y)

    scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=80, edgecolors='k', facecolors='none', zorder=10)
    scatter(X[:, 0], X[:, 1], c=Y, zorder=10, edgecolors='k', cmap=plt.cm.Paired)

    axis('tight')
    x_min = -3
    x_max = 3
    y_min = -3
    y_max = 3

    XX, YY = np.mgrid[x_min:x_max:200j, y_min:y_max:200j]
    Z = clf.decision_function(np.c_[XX.ravel(), YY.ravel()])

    Z = Z.reshape(XX.shape)
    pcolormesh(XX, YY, Z > 0, cmap=plt.cm.Paired)
    contour(XX, YY, Z, colors=['k', 'k', 'k'], linestyles=['--', '-', '--'], levels=[-.5, 0, .5])

    xlim(x_min, x_max)
    ylim(y_min, y_max)
    
    title('Method: {0}'.format(kernel))
    
    show()
```
